**1. Import c√°c th∆∞ vi·ªán**

```python
import pandas as pd
import numpy as np
import random
```

**2. C√°c ng√†nh ngh·ªÅ c∆° b·∫£n**

```python
JOBS = [
    "IT Engineer", "Data Analyst", "Teacher", "Nurse", "Accountant",
    "Salesperson", "Freelancer", "Designer", "Mechanic", "Student"
]
```

**3. C√°c h√†m sinh d·ªØ li·ªáu**

```python
# Sinh d·ªØ li·ªáu theo ngh·ªÅ
def generate_income(job):
    if job in ["IT Engineer", "Data Analyst", "Accountant", "Designer"]:
        return np.random.randint(12000000, 30000000)
    elif job in ["Teacher", "Nurse", "Mechanic", "Salesperson"]:
        return np.random.randint(6000000, 15000000)
    else:  # Student, Freelancer
        return np.random.randint(3000000, 12000000)

# 2. Chi ti√™u
def generate_expense(income):
    ratio = np.random.uniform(0.45, 0.95)
    return int(income * ratio)

# 3. N·ª£
def generate_debt(income):
    level = np.random.choice(["low", "medium", "high"], p=[0.5, 0.3, 0.2])
    if level == "low":
        return np.random.randint(0, int(income * 0.3))
    if level == "medium":
        return np.random.randint(int(income * 0.3), int(income * 0.8))
    return np.random.randint(int(income * 0.8), int(income * 1.2))

# 4. L·ªãch s·ª≠ tr·∫£ ch·∫≠m
def generate_late_payments(debt):
    if debt < 3000000:
        return np.random.choice([0, 1])
    elif debt < 10000000:
        return np.random.choice([0, 1, 2])
    else:
        return np.random.choice([1, 2, 3, 4])


# 5. FICO: chi·ªÅu d√†i l·ªãch s·ª≠ t√≠n d·ª•ng
def generate_history_length(job):
    if job == "Student":
        return np.random.randint(0, 3)
    return np.random.randint(1, 15)


# 6. T√†i kho·∫£n t√≠n d·ª•ng m·ªõi
def generate_new_accounts():
    return np.random.choice([0, 1, 2, 3])


# 7. Credit mix
CREDIT_MIX_TYPES = ["good", "fair", "poor"]
def generate_credit_mix():
    return random.choice(CREDIT_MIX_TYPES)


# 8. T√≠nh ƒëi·ªÉm FICO gi·∫£ l·∫≠p
def calculate_credit_score(income, expense, debt, late, history, new_acct, mix):
    score = 300

    # 35% l·ªãch s·ª≠ tr·∫£ n·ª£
    score += max(0, 200 - late * 40)

    # 30% debt ratio
    debt_ratio = debt / max(income, 1)
    if debt_ratio < 0.3:
        score += 150
    elif debt_ratio < 0.6:
        score += 70
    else:
        score -= 50

    # 15% l·ªãch s·ª≠ t√≠n d·ª•ng
    if history >= 10:
        score += 80
    elif history >= 5:
        score += 40
    else:
        score += 10

    # 10% t√†i kho·∫£n t√≠n d·ª•ng m·ªõi
    score -= new_acct * 15

    # 10% credit mix
    if mix == "good":
        score += 40
    elif mix == "fair":
        score += 20

    return max(300, min(score, 850))
```

**4. Sinh 150 m·∫´u d·ªØ li·ªáu**

```python
rows = []
for i in range(1, 1000):
    job = random.choice(JOBS)
    income = generate_income(job)
    expense = generate_expense(income)
    debt = generate_debt(income)
    late = generate_late_payments(debt)
    history = generate_history_length(job)
    new_account = generate_new_accounts()
    mix = generate_credit_mix()

    credit_score = calculate_credit_score(income, expense, debt, late, history, new_account, mix)

    rows.append({
        "id": id,
        "job": job,
        "income_monthly": income,
        "expense_monthly": expense,
        "debt_amount": debt,
        "late_payments_12m": late,
        "credit_history_length_years": history,
        "new_credit_accounts": new_account,
        "credit_mix": mix,
        "income_expense_ratio": round(income / expense, 2),
        "debt_ratio": round(debt / income, 2),
        "credit_score": credit_score,
        "credit_score_class": "good" if credit_score >= 750 else ("fair" if credit_score >= 650 else "bad")
    })
```
